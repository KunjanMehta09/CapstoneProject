```python
dematel_weights=[] # Initializaing list to store final DEMATEL weights
flag=0
for i in criteria.values():
    flag+=1
    X=[]
    a=np.array([[[0,0,0],0],[0,[0,0,0]]])
    l1=input(f'Enter ratings of {i[0]} wrt {i[1]}: ') # Taking user input for Sub-criteria v/s Sub-criteria sets
    l2=input(f'Enter ratings of {i[1]} wrt {i[0]}: ')
    a[0][1]=list(lin_term_dematel[l1.upper()])
    a[1][0]=list(lin_term_dematel[l2.upper()]) # Obtaining Fuzzy Direct Relation Matrix
    s=max(a[0][1][2],a[1][0][2]) # Performing computation
    a[0][1]=np.divide(a[0][1],s)
    a[1][0]=np.divide(a[1][0],s)
    l=np.array([a[i][j][0] for i in range(2) for j in range(2)])
    m=np.array([a[i][j][1] for i in range(2) for j in range(2)])
    u=np.array([a[i][j][2] for i in range(2) for j in range(2)])
    l=l.reshape(2,2)
    m=m.reshape(2,2)
    u=u.reshape(2,2)
    I_l,I_m,I_u=np.subtract(I,l),np.subtract(I,m),np.subtract(I,u)
    inv_l,inv_m,inv_u=np.linalg.inv(I_l),np.linalg.inv(I_m),np.linalg.inv(I_u)
    print(inv_l)
    print(inv_m)
    print(inv_u)
    l1,m1,u1=l.dot(inv_l),m.dot(inv_m),u.dot(inv_u)
    t_m1,t_m2=[],[]
    t_m1=[[l1[0][0],m1[0][0],u1[0][0]],[l1[1][0],m1[1][0],u1[1][0]]]
    t_m2=[[l1[0][1],m1[0][1],u1[0][1]],[l1[1][1],m1[1][1],u1[1][1]]]
    d1,d2=[],[]
    for j in range(2):
        d1.append((t_m1[j][0]+4*t_m1[j][1]+t_m1[j][2])/6)
        d2.append((t_m2[j][0]+4*t_m2[j][1]+t_m2[j][2])/6)
    d1.append(d1[0]+d1[1])
    d2.append(d2[0]+d2[1])
    inner=[[0,0],[0,0]]
    for j in range(2):
        inner[j]=[d1[j]/d1[2],d2[j]/d2[2]] # Obtain final DEMATEL Weights for each Sub-Criteria set
    dematel_weights.append(inner)
```

# Fuzzy ANP

```python
pairwise_matrix=np.ones([6,6],dtype=int) # Initializing Matrix
main_criteria=list(criteria.keys()) # Implementing fuzzy ANP for Criteria v/s Criteria, i.e. Outer Dependancy
pairwise_matrix=pd.DataFrame(pairwise_matrix,columns=main_criteria,index=main_criteria,dtype=object) # Creating Dataframe for pairwise comparision matrix
for j in range(6): # Reading user input
    for k in range(6):
        if(j==k):
            x=(1,1,1)
            pairwise_matrix[main_criteria[k]][main_criteria[j]]=x
        if(k>j):
            l=int(input(f'Enter rating {main_criteria[j]} wrt {main_criteria[k]}: '))-1
            x=lin_term_anp[l]
            pairwise_matrix[main_criteria[k]][main_criteria[j]]=x
            pairwise_matrix[main_criteria[j]][main_criteria[k]]=(1/x[2],1/x[1],1/x[0])
        
pairwise_matrix
```

```python
opt=0.5 # Defining Degree Optimism
matrix2=pairwise_matrix.copy()
for j in range(6): # Performing ANP Computations for Criteria v/s Criteria
    for k in range(6):
        x=matrix2[main_criteria[k]][main_criteria[j]]
        w=opt*x[2]+(1-opt)*x[0]
        matrix2[main_criteria[k]][main_criteria[j]]=w
total=list(matrix2.sum(axis=0)) 
matrix2
nor_matrix=matrix2.copy() # Pairwise Comparison Matrix for the Relative Importance of Criteria
nor_matrix
```

```python
nor_matrix=matrix2.copy()
max_eigen=0 # Initializing maximum eigen value
for j in range(6):
    for k in range(6):
        x=matrix2[main_criteria[k]][main_criteria[j]]/total[k]
        nor_matrix[main_criteria[k]][main_criteria[j]]=x
priority_vector=1/6*nor_matrix.sum(axis=1)
for j in range(len(total)):
    max_eigen+=total[j]*priority_vector[j] # Finding maximum eigen value
ci=(max_eigen-6)/5 # Calculating Consistency Index
cr=ci/1.24 # Defining Consistency Ratio
priority_vector=list(priority_vector)
nor_matrix # Normalized Relative Weights
```

```python
def wsc(l): # Defining functions for Fuzzy VIKOR Calculations
    minimum=1
    for j in l:
        x=j[0]
        if(x<minimum):
            minimum=x
    maximum=0
    for j in l:
        x=j[2]
        if(x>maximum):
            maximum=x
    average,x=0,0
    for j in l:
        x+=j[1]
    average=x/len(l)
    result=[minimum,average,maximum]
    return(result)

def min_value(l):
    result=[]
    for i in range(0,3):
        minimum=10
        for j in l:
            x=j[i]
            if(x<minimum):
                minimum=x
        result.append(minimum)
    return(result)

def max_value(l):
    result=[]
    for i in range(0,3):
        maximum=0
        for j in l:
            x=j[i]
            if(x>maximum):
                maximum=x
        result.append(maximum)
    return(result)
def da_sc(l1,l2):
    da_sc=[]
    dif=l2[0][2]-l2[1][0]
    for j in range(len(l1)):
        x=[]
        for k in range(3):
            x.append((l2[0][k]-l1[j][-(k+1)])/dif)
        da_sc.append(x)
    return(da_sc)
def w_da(l1,l2):
    result=[]
    for j in l2:
        for k in range(len(l1)):
            result.append(l1[k]*j[k])
    return(result)
```

```python
vikor_weights=[] # Reading input for VIKOR and performing computation in order to obtain VIKOR weights
for i in criteria.values():
    val1,val2,wsc1,wsc2=[],[],[],[]
    l1=list(input(f'Enter ratings by decision makers wrt {i[0]}: ').split())
    l2=list(input(f'Enter ratings by decision makers wrt {i[1]}: ').split())
    for j in range(len(l1)):
        val1.append(lin_term_vikor[l1[j].upper()])
    for j in range(len(l2)):
        val2.append(lin_term_vikor[l2[j].upper()])
    wsc1=wsc(val1)
    wsc2=wsc(val2)
    l1,l2=[],[]
    for j in alternatives:
        x=int(input(f'Enter ratings of {i[0]} with respect to {j}: '))-1
        l1.append(lin_term_anp[x])
        x=int(input(f'Enter ratings of {i[1]} with respect to {j}: '))-1
        l2.append(lin_term_anp[x])
    sc1_f,sc2_f=[],[]
    sc1_f.append(list(max_value(l1)))
    sc1_f.append(list(min_value(l1)))
    sc2_f.append(list(max_value(l2)))
    sc2_f.append(list(min_value(l2)))
    da_sc1=da_sc(l1,sc1_f)
    da_sc2=da_sc(l2,sc2_f)
    w_da1=w_da(wsc1,da_sc1)
    w_da2=w_da(wsc2,da_sc2)
    s_j,r_j=[],[]
    for j in range(len(w_da1)):
        s_j.append(w_da1[j]+w_da2[j])
        r_j.append(max(w_da1[j],w_da2[j])) 
    s_star,r_star=[],[]
    for j in range(3):
        x1,x2=[],[]
        k=j
        while(k<len(s_j)):
            x1.append(s_j[k])
            x2.append(r_j[k])
            k+=3
        if(j==2):
            s_oc=max(x1)
            r_oc=max(x2)
        s_star.append(min(x1))
        r_star.append(min(x2))
    q=[]
    for j in range(len(s_j)):
        x=0.5*(s_j[j]-s_star[-(j%3+1)])/(s_oc-s_star[0])+(1-0.5*(r_j[j]-r_star[-(j%3+1)])/(r_oc-r_star[0]))
        q.append(round(x,4))
    defuzzified_weights=[]
    j=0
    while(j<len(q)):
        x=(q[j]+4*q[j+1]+q[j+2])/6
        defuzzified_weights.append(round(x,4))
        j+=3
    vikor_weights.append(defuzzified_weights)
vikor_weights   
```
